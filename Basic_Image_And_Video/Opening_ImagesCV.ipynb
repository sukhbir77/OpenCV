```python
#imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
#Import images as a array.
img = cv2.imread('Data/00-puppy.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
# Matplotlib --> RGB RED GREEN BLUE
# OpenCV --> Blue Green Red
```

```python
# To convert the right colors.
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img_gray = cv2.imread('Data/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray, cmap = 'gray')
img_gray.shape
```

```python
#Resize the image.
img = cv2.resize(img,(200,200))
plt.imshow(img)
```

```python
w_ratio  = 0.5
h_ratio = 0.5
```

```python
img = cv2.resize(img_gray, (0,0), img_gray, w_ratio, h_ratio)
```

```python
plt.imshow(img)
```

```python
#Flip the images.
```

```python
img = cv2.flip(img, -1)
plt.imshow(img)
```

```python
#write the image to another files.
cv2.imwrite('New_Image.jpg', img)
```

```python
#Display the large image
fig = plt.figure(figsize=(20,8))
ax = fig.add_subplot(111)
ax.imshow(img)
```
