```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Data/00-puppy.jpg')
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#Convert to HSV
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img)
```
